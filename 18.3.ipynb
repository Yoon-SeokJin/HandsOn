```python
import gym

env = gym.make('CartPole-v0')
obs = env.reset()

answer = 0
for i in range(500):
    env.render()
    if obs[2] > 0:
        action=1
    else:
        action=0
    obs, reward, done, info = env.step(action)
    answer += reward
    print(answer)

env.close()
print(answer)
```
